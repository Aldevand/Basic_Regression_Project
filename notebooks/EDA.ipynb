```python
# import library
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import os
print("seluruh pustaka berhasil dimuat")
```

```python
# buat jalur untuk baca data os.path.join("..", "data")

folder = os.path.join("..", "data")
file = "Data_Preprocessing.csv"
path = os.path.join(folder, file)
print("jalur berhasil dibuat", {path})

#  baca data
df = pd.read_csv(path)
df.head(5)
```

```python
df.info()
```

**Line Plot**

```python
# buat line plot -> sumbu X (waktu) dan sumbu Y (Total)
plt.figure(figsize=(10, 5)) #atur ukuran grafik

waktu = df["Waktu"]
total = df["Total"]
plt.plot(waktu, total, marker='o', markersize=5, linestyle='-', color='blue')

# judul grafik
plt.title("Jumlah Penggangguran Per Tahun", fontsize=14)
plt.xlabel("Waktu", fontsize=12)
plt.ylabel("Total", fontsize=12)

# rotasi label tahun
plt.xticks(waktu, rotation=45)

# tambah grid
plt.grid(True, linestyle="--", alpha=0.5)

plt.tight_layout()
plt.show()
```

```python
# ubah tipe data waktu -> datetime
df["Waktu"] = pd.to_datetime(df["Waktu"])
print("konversi Waktu berhasil")

# ambil bagian tahun di kolom Waktu
df["Tahun"] = df["Waktu"].dt.year
df.head(3)
```

```python
df_total = df.groupby("Tahun")["Total"].mean().reset_index()

plt.figure(figsize=(12, 6))
plt.plot(df_total["Tahun"], df_total["Total"], marker='o', linestyle='-', color='blue')

plt.title("Tren Total Pengangguran per Tahun", fontsize=14)
plt.xlabel("Tahun", fontsize=12)
plt.ylabel("Total Pengangguran", fontsize=12)
plt.xticks(df_total["Tahun"], rotation=45)
plt.grid(True, linestyle='--', alpha=0.5)
plt.tight_layout()
plt.show()
```

##  Visualisasi Tren Total Pengangguran per Tahun

Grafik berikut menampilkan **tren jumlah total pengangguran di Indonesia** dari tahun 2006 hingga 2022. Data diambil dua kali per tahun (Februari dan Agustus), lalu dirata-ratakan agar grafik lebih smooth dan mudah dibaca.

###  Detail Visualisasi:
- **Sumbu X**: Tahun (20062022)
- **Sumbu Y**: Total pengangguran (dalam satuan juta orang)
- **Garis**: Menunjukkan perubahan jumlah pengangguran dari tahun ke tahun
- **Titik**: Menandai nilai rata-rata pengangguran tiap tahun

###  Insight Awal:
- Pengangguran tertinggi terjadi di awal tahun 2006
- Terjadi penurunan signifikan hingga sekitar 2012
- Periode 20122019 relatif stabil
- Lonjakan tajam terjadi pada tahun 2020 (dampak pandemi COVID-19)
- Penurunan kembali terjadi di 2021 dan 2022, meskipun belum ke titik terendah

```python
# hitung proporsi tiap jenjang -> total
jenjang = [
    "Tidak/belum pernah sekolah",
    "Tidak/belum tamat SD",
    "SD",
    "SLTP",
    "SLTA Umum/SMU",
    "SLTA Kejuruan/SMK",
    "Akademi/Diploma",
    "Universitas"
]
df_proporsi = df[jenjang].div(df["Total"], axis=0)
```

```python
df_proporsi
```

```python
# Hitung rata-rata proporsi tiap jenjang (keseluruhan periode)
mean_proporsi = df_proporsi.mean().sort_values(ascending=False)
```

```python
mean_proporsi
```

```python
plt.figure(figsize=(10, 5))
mean_proporsi.plot(kind="bar", color="red")

plt.title("Presentase jumlah pengangguran tiap jenjang", fontsize=12)
plt.ylabel("Proporsi terhadap Total Pengangguran", fontsize=12)
plt.xticks(rotation=45)
plt.grid(axis="y", linestyle="--", alpha=0.5)
plt.tight_layout()
plt.show()
```

#  Analisis Proporsi Pengangguran Berdasarkan Jenjang Pendidikan

## 1.  Tujuan Analisis
Mengetahui kontribusi masing-masing jenjang pendidikan terhadap total angka pengangguran, sehingga dapat digunakan sebagai dasar pertimbangan dalam perumusan kebijakan ketenagakerjaan dan pendidikan.

---

## 2.  Proses Perhitungan

### a. Normalisasi Data
- Data awal berisi jumlah pengangguran per jenjang pendidikan untuk setiap tahun.
- Dilakukan **normalisasi** dengan membagi jumlah pengangguran tiap jenjang terhadap total pengangguran per tahun.
- Hasilnya adalah **proporsi tahunan** untuk masing-masing jenjang pendidikan.

### b. Agregasi Rata-rata
- Proporsi tahunan dirata-ratakan menggunakan fungsi `.mean()` untuk mendapatkan **proporsi rata-rata** tiap jenjang selama periode data.
- Hasil kemudian diurutkan dari yang tertinggi ke terendah untuk memudahkan interpretasi.

###  Interpretasi Visualisasi Proporsi Pengangguran

Diagram batang menunjukkan distribusi proporsi pengangguran berdasarkan jenjang pendidikan. Beberapa poin penting yang dapat disimpulkan:

- **SMA Umum/SMU** menempati posisi tertinggi dalam proporsi pengangguran (~26%), menunjukkan bahwa lulusan pendidikan menengah umum paling rentan terhadap pengangguran.
- **SLTP dan SMK** juga memiliki proporsi signifikan, menandakan tantangan transisi dari pendidikan menengah ke dunia kerja.
- **Lulusan Universitas dan Diploma** memiliki proporsi lebih rendah, mengindikasikan bahwa pendidikan tinggi cenderung memberikan peluang kerja lebih baik.
- **Kelompok tanpa pendidikan formal atau belum tamat SD** memiliki proporsi terendah, kemungkinan karena jumlahnya kecil atau tidak tercatat secara formal dalam data.

> Hasil ini dapat menjadi dasar bagi perumusan kebijakan ketenagakerjaan dan pendidikan, khususnya dalam meningkatkan kesiapan kerja lulusan SMA dan SMK.



              ** Pearson Corelation **

```python
# buat jalur untuk baca data
folder = os.path.join("..", "data")
file = "Data_Preprocessing.csv"
path = os.path.join(folder, file)
print("jalur berhasil dibuat", {path})

#  baca data
df = pd.read_csv(path)
df.head(5)
```

```python
# hapus kolom waktu dan hitung korelasi antar variabel lainnya
corr = df.drop(columns="Waktu").corr(method="pearson")
corr.head(5)
```

```python
# visualisasi heatmap-nya
plt.figure(figsize=(10, 5))
sns.heatmap(corr, annot=True, cmap="coolwarm", fmt=".2f")
plt.title("Korelasi antar variabel")
plt.show()
```

#  Analisis Korelasi Antar Jenjang Pendidikan

## 1.  Metode Analisis
Analisis korelasi dilakukan menggunakan **metode Pearson correlation** untuk mengukur hubungan linear antar jenjang pendidikan berdasarkan jumlah pengangguran.  
Nilai korelasi berada pada rentang **-1 hingga +1**, di mana:
- Nilai mendekati **+1** menunjukkan hubungan positif kuat.
- Nilai mendekati **-1** menunjukkan hubungan negatif kuat.
- Nilai mendekati **0** menunjukkan hubungan yang lemah atau tidak linear.

---

## 2.  Penghapusan Variabel Waktu
Kolom **`Waktu`** dihapus dari dataset karena tidak relevan dalam analisis hubungan antar jenjang pendidikan.  
Tujuan analisis ini adalah untuk melihat **hubungan antar kategori pendidikan secara struktural**, bukan perubahan nilainya terhadap waktu.

---

## 3.  Interpretasi Visualisasi
Berdasarkan heatmap hasil korelasi:

- Korelasi **tinggi dan positif** terjadi antara jenjang **SD**, **SLTP**, dan **SLTA Umum/SMU**, menandakan jumlah pengangguran di jenjang ini cenderung bergerak bersama.
- **SLTA Kejuruan/SMK** memiliki korelasi **negatif** terhadap jenjang dasar dan menengah umum, menunjukkan pola pergerakan yang berlawanan.
- **Universitas** juga memiliki korelasi negatif dengan jenjang rendah, mengindikasikan tren pengangguran yang berbeda arah.
- Korelasi positif yang kuat antara beberapa jenjang dan kolom **Total** menunjukkan kontribusi besar jenjang tersebut terhadap total pengangguran.

---

## 4.  Kesimpulan
Analisis korelasi Pearson menunjukkan adanya keterkaitan kuat antar jenjang pendidikan menengah dan dasar, sementara jenjang tinggi (Diploma dan Universitas) memperlihatkan pola yang berbeda.  
Temuan ini dapat menjadi dasar untuk memahami hubungan struktural antar tingkat pendidikan dalam konteks pengangguran.
