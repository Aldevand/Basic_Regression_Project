```python
import pandas as pd
import numpy as np
import os
print("pustaka berhasil dimuat")
```

```python
# baca dataset
folder = os.path.join("..", "data")
file = "Data_Preprocessing1.csv"
path = os.path.join(folder, file)

#baca data
df = pd.read_csv(path)
df.head(5)
```

```python
df.info()
```

```python
# konversi tipe data waktu ->datetime
df["Waktu"] = pd.to_datetime(df["Waktu"])
df.info()
```

```python
# pisahkan bagian Tahun jadi fitur/kolom baru
df["Tahun"] = df["Waktu"].dt.year
# cek hasilnya
df[["Waktu", "Tahun"]].head(5)
```

```python
# buat Fitur Lag
df["Total_lag1"] = df["Tahun"].shift(1)
# cek hasilnya
df[["Waktu","Tahun","Total_lag1"]].head(5)
```

```python
# buat Fitur Pertumbuhan Tahunan (Growth Rate)
df["Growth_Rate"] = df["Total"].pct_change() * 100
# cek hasilnya 
df[["Tahun", "Total", "Growth_Rate"]].head(10)
```

```python
df
```

```python
# Hapus baris pertama (karena tidak punya data tahun sebelumnya)
df = df.dropna(subset=["Total_lag1", "Growth_Rate"]).reset_index(drop=True)

# Cek kembali dimensi data
df.info()
```
